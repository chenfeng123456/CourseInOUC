```python
def create_gif(img_list, gif_name, duration):
    frames = []
    for img in  img_list:
        frames.append(imageio.imread(img))
    imageio.mimsave(gif_name+'.gif', frames, 'GIF', duration=duration)
    return imageio.mimread('%s.gif' % gif_name)
        
```

```python
img_list = [ '01.jpg', '02.jpg', '03.jpg']
my_gif = create_gif(img_list, 'myGIF', 1)
wade_and_james = imageio.mimread('wade_and_james.gif')
print( type( my_gif ))
for i in my_gif:
    print( type(i))
```

```python
#https://www.cnblogs.com/yucen/p/9343554.html

#frames = [f.copy() for f in ImageSequence.Iterator(im)]
with Image.open('wade_and_james.gif') as im:
    r_wade_and_james = [ f.copy() for f in ImageSequence.Iterator(im) ]
    r_wade_and_james.reverse()
    r_wade_and_james[0].save('r_wade_and_james.gif',save_all=True, append_images=r_wade_and_james[1:])
    
with Image.open('wade_and_james.gif') as im:
    if im.is_animated:
        print("is_animated")
        frames = [f.copy() for f in ImageSequence.Iterator(im)]
        frames.reverse() # 内置列表倒序方法
        # 将倒序后的所有帧图像保存下来
        frames[0].save('wade_and_james_reverse.gif', save_all=True, append_images=frames[1:])
```
